```python
from ultralytics import YOLO
import os
import shutil
```

```python
# –¢–∏–ø—ã –¥–æ—Å—Ç—É–ø–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π Yolo
models = [
    "yolo11n-cls",  # 0 - nano-–º–æ–¥–µ–ª—å
    "yolo11m-cls",  # 1 - medium-–º–æ–¥–µ–ª—å
    "yolo11x-cls"  # 2 - large-–º–æ–¥–µ–ª—å
]

# –ò–Ω–¥–µ–∫—Å –≤—ã–±—Ä–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ (—Å–º. models)
type_model = 0

# –ü—É—Ç—å –¥–æ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏, —Å–æ–¥–µ—Ä–∂–∞—â–∏–π –≤–ª–æ–∂–µ–Ω–Ω—ã—Ö –ø–æ–¥–¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏ —Å –¥–∞–Ω–Ω—ã–º–∏:
# /train
# /test
# /val (–æ–ø—Ü–∏–æ–Ω–∞–ª—å–Ω–æ)
path_to_data = "./images/"

# –ó–∞–≥—Ä—É–∑–∫–∞ –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏
model = YOLO(f"./pretrained_models/{models[type_model]}.pt")
```

```python
# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
results = model.train(
    #resume=True,  # –ü—Ä–æ–¥–æ–ª–∂–µ–Ω–∏–µ –¥–æ–æ–±—É—á–µ–Ω–∏—è –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏
    data=path_to_data,  # –î–∏—Ä–µ–∫—Ç–æ—Ä–∏—è —Å –¥–∞–Ω–Ω—ã–º–∏
    imgsz=640,
    epochs=10,
    batch=8,
    # –î–∏—Ä–µ–∫—Ç–æ—Ä–∏—è –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    name=f"{models[type_model]}_custom"
)
```

```python
# –í–∞–ª–∏–¥–∞—Ü–∏—è –º–æ–¥–µ–ª–∏
# –ü—Ä–æ–≤–µ—Ä–∫–∞: –î–∏—Ä–µ–∫—Ç–æ—Ä–∏—è —Å –¥–∞–Ω–Ω—ã–º–∏ /val —Å—É—â–µ—Å—Ç–≤—É–µ—Ç?
if os.path.isdir(f"{path_to_data}/val"):
    metrics = model.val()  # –ê—Ä–≥—É–º–µ–Ω—Ç—ã –Ω–µ —Ç—Ä–µ–±—É—é—Ç—Å—è
    metrics	.top1  # –¢–æ–ø1 accuracy
    metrics.top5  # –¢–æ–ø5 accuracy
```

```python
# –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –º–æ–¥–µ–ª–∏
%matplotlib
import matplotlib.pyplot as plt
from skimage import io
from pathlib import Path
import re


# –î–ª—è –∫–∞–∂–¥–æ–≥–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –æ—Ç–æ–±—Ä–∞–∂–∞–µ—Ç—Å—è:
# –°–∞–º–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ.
# –ò—Å—Ç–∏–Ω–Ω—ã–π –∫–ª–∞—Å—Å (True Label).
# –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–π –∫–ª–∞—Å—Å (Predicted Label).
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π —Å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è–º–∏
def visualize_predictions(model, path_to_data, num_images=5):

    # –í–∑—è—Ç–∏–µ –ø—É—Ç–∏ –¥–æ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏ /test
    path_to_test = Path(f"{path_to_data}/test")
    
    # –ò–º–ø–æ—Ä—Ç –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –∏–∑ /test –∏ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏—Ö –∫–ª–∞—Å—Å–∞
    imgs = []  # –°–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
    true_classes = []  # –°–ø–∏—Å–æ–∫ –¥–ª—è –∏—Å—Ç–∏–Ω–Ω—ã—Ö –∫–ª–∞—Å—Å–æ–≤
    # –ü–µ—Ä–µ–±–æ—Ä –≤—Å–µ—Ö –≤–ª–æ–∂–µ–Ω–Ω—ã—Ö –ø—É—Ç–µ–π
    for path_to_image in path_to_test.rglob("*"):
        if len(imgs) >= num_images:
            break
        base, ext = os.path.splitext(path_to_image)  # –í–∑—è—Ç–∏–µ —Ä–∞—Å—à–∏—Ä–µ–Ω–∏—è
        # –ü—Ä–æ–≤–µ—Ä–∫–∞: –¥–ª–∏–Ω–∞ —Ä–∞—Å—à–∏—Ä–µ–Ω–∏—è –Ω–µ –±–æ–ª–µ–µ 5 —Å–∏–º–≤–æ–ª–æ–≤ (—Å —É—á–µ—Ç–æ–º —Ç–æ—á–∫–∏)?
        if ext != '' and len(ext) <= 5:
            try:
                img = io.imread(path_to_image)  # –ò–º–ø–æ—Ä—Ç –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
                # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –≤ —Å–ø–∏—Å–æ–∫
                imgs.append(img)
                # –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞
                regex = re.search(r"/[^test]*/", base)[0]
                true_class = regex.replace('/', '')
                true_classes.append(true_class)
            except:
                continue

    num_images = len(imgs)  # –ü–æ–¥—Å—á–µ—Ç —É—Å–ø–µ—à–Ω–æ –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π

    # –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
    results = model(imgs)
    
    # –í—ã—á–∏—Å–ª—è–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –∏ —Å—Ç–æ–ª–±—Ü–æ–≤
    num_rows = (num_images + 1) // 2  # –ú–∞–∫—Å–∏–º—É–º 2 –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –≤ —Å—Ç—Ä–æ–∫—É
    num_cols = min(num_images, 2)    # –ù–µ –±–æ–ª—å—à–µ 2 —Å—Ç–æ–ª–±—Ü–æ–≤

    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(10, 5 * num_rows))
    if num_images == 1:
        axes = np.array([axes])  # –ß—Ç–æ–±—ã axes –±—ã–ª –º–∞—Å—Å–∏–≤–æ–º, –¥–∞–∂–µ –µ—Å–ª–∏ num_images = 1

    axes = axes.ravel()  # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º axes –≤ –æ–¥–Ω–æ–º–µ—Ä–Ω—ã–π –º–∞—Å—Å–∏–≤ –¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞

    for i, image in enumerate(imgs):
        ax = axes[i]
        ax.imshow(image)
        ax.set_title(f"1")

        # –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞
        result = results[i]  # —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–ª—è —Ç–µ–∫—É—â–µ–≥–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è 
        max_prob = max(result.probs.data)  # –ü–æ–∏—Å–∫ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
        # –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏–Ω–¥–µ–∫—Å–∞ –Ω–∞–∏–±–æ–ª–µ–µ –≤–µ—Ä–æ—è—Ç–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞
        idx_max_prob = list(result.probs.data).index(max_prob)
        predicted_class = result.names[idx_max_prob]  # –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–π –∫–ª–∞—Å—Å

        true_class = true_classes[i]  # –ò—Å—Ç–∏–Ω–Ω—ã–π –∫–ª–∞—Å—Å
        
        ax.set_title(
            f"True: {true_class}"
            + f"\nPred: {predicted_class}", 
            fontsize=10
        )
        ax.axis('off')  # –°–∫—Ä—ã–≤–∞–µ–º –æ—Å–∏

    # –°–∫—Ä—ã–≤–∞–µ–º –ø—É—Å—Ç—ã–µ subplots, –µ—Å–ª–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –Ω–µ –∫—Ä–∞—Ç–Ω–æ 2
    for j in range(num_images, num_rows * num_cols):
        axes[j].axis('off')

    plt.tight_layout()  # –£–ª—É—á—à–∞–µ–º —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏–µ subplots
    plt.show()


# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è
visualize_predictions(model, path_to_data, num_images=9)
```
